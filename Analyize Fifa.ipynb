#### What are the top 10 clubs in the wage and visualize the results ?

```python
top10 = df.groupby('Club')['Wage'].sum().sort_values(ascending = False).head(10)
top10
```

```python
top_10 = df[df['Club'].isin(top10.index)]
top_10
```

```python
sns.barplot(data = top_10 , y = 'Club' , x = 'Wage' ,
            estimator = np.sum , order = top10.index )
```

#### is there any outlier in the Age find it with 1- visualization 2-  with pandas ?

```python
sns.boxplot(data = df , x = 'Age' )
```

```python
ag_sts = df['Age'].describe()
ag_sts
```

```python
IQR = ag_sts['75%'] - ag_sts['25%']
max_wisker = ag_sts['75%'] + (1.5 * IQR)
min_wisker = ag_sts['25%'] - (1.5 * IQR)
```

```python
outlier = df[ (df['Age'] > max_wisker) | (df['Age'] < min_wisker)]
```

```python
outlier
```

```python
sns.violinplot(data = df  , y = 'Age' )
```

#### Is there a correlation between Value and Overall and viusalize the results with two graphs?**

```python
df['Value'].corr(df['Overall']) # weak +ve correlation
```

```python
cor = df.corr()[['Value' , 'Overall']].loc[['Value' , 'Overall']]
cor
```

```python
sns.scatterplot(data = df , x = 'Value' , y = 'Overall')
```

#### what are the ratio of the Preferred Foot and visualize with a graph ?

```python
no = df['Preferred Foot'].value_counts()
ratio = no * 100 / df['Preferred Foot'].value_counts().sum()
ratio
```

```python
sns.countplot(data = df , x = 'Preferred Foot')
```

```python
sns.barplot(data = ratio.to_frame() , x = ratio.index , y = ratio)
```

#### what are the top 10 average overall clubs and visualize with a graph ?

```python
top10 = df.groupby('Club')['Overall'].mean().sort_values(ascending = False).head(10)
top10
```

```python
top_10 = df[df['Club'].isin(top10.index)]
```

```python
sns.barplot(data = top_10 , y = 'Club' , x = 'Overall' , estimator = np.mean , order = top10.index)
```
